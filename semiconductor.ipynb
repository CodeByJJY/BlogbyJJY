# **Comparing**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Example data: 3 datasets
data_set_1 = [
    9.163E-9, 9.179E-9, 9.194E-9, 9.207E-9, 9.221E-9, 9.233E-9, 9.245E-9,
    9.257E-9, 9.266E-9, 9.276E-9, 9.285E-9, 9.293E-9, 9.299E-9, 9.305E-9,
    9.309E-9, 9.312E-9, 9.315E-9, 9.317E-9, 9.318E-9, 9.318E-9, 9.318E-9,
    9.317E-9, 9.313E-9, 9.311E-9, 9.308E-9
]
data_set_2 = [
    10.96E-9, 10.99E-9, 11.01E-9, 11.04E-9, 11.06E-9, 11.08E-9, 11.10E-9,
    11.12E-9, 11.14E-9, 11.15E-9, 11.16E-9, 11.18E-9, 11.19E-9, 11.20E-9,
    11.20E-9, 11.21E-9, 11.21E-9, 11.21E-9, 11.21E-9, 11.22E-9, 11.22E-9,
    11.21E-9, 11.21E-9, 11.20E-9, 11.19E-9
]
data_set_3 = [
    6.227E-9, 6.255E-9, 6.283E-9, 6.308E-9, 6.334E-9, 6.359E-9, 6.384E-9,
    6.407E-9, 6.430E-9, 6.452E-9, 6.473E-9, 6.494E-9, 6.513E-9, 6.532E-9,
    6.549E-9, 6.567E-9, 6.583E-9, 6.599E-9, 6.613E-9, 6.627E-9, 6.640E-9,
    6.653E-9, 6.664E-9, 6.675E-9, 6.684E-9
]

# Function to calculate normalized x-axis and PDF for a dataset
def normalize_and_calculate_pdf(data, ref_mean):
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = [(x - mean + ref_mean) for x in data]
    x = np.linspace(min(normalized_data), max(normalized_data), 1000)
    y = norm.pdf(x, ref_mean, std_dev)
    return normalized_data, x, y, mean, std_dev

# Reference mean (arbitrarily chosen as the mean of data_set_1)
ref_mean = np.mean(data_set_1)

# Calculate normalized PDFs for all datasets
normalized_1, x1, y1, mean1, std1 = normalize_and_calculate_pdf(data_set_1, ref_mean)
normalized_2, x2, y2, mean2, std2 = normalize_and_calculate_pdf(data_set_2, ref_mean)
normalized_3, x3, y3, mean3, std3 = normalize_and_calculate_pdf(data_set_3, ref_mean)

# Plotting the graph
plt.figure(figsize=(12, 8))

# Plotting the normalized distributions
plt.plot(x1, y1, label=f'Addition-Based (μ={mean1:.2e}, σ={std1:.2e})', color='blue', linewidth=2)
plt.plot(x2, y2, label=f'Current-served (μ={mean2:.2e}, σ={std2:.2e})', color='green', linewidth=2)
plt.plot(x3, y3, label=f'Basic Ring Oscillator (μ={mean3:.2e}, σ={std3:.2e})', color='red', linewidth=2)

# Scatter plots for the original data points
plt.scatter(normalized_1, [0] * len(normalized_1), color='blue', label='Data Points 1 (Scatter)', alpha=0.6)
plt.scatter(normalized_2, [0] * len(normalized_2), color='green', label='Data Points 2 (Scatter)', alpha=0.6)
plt.scatter(normalized_3, [0] * len(normalized_3), color='red', label='Data Points 3 (Scatter)', alpha=0.6)

# Add labels, title, and legend
plt.title('Normal Distributions Comparing', fontsize=16)
plt.xlabel('Timing Delay (s, Normalized)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Display the graph
plt.show()
```
